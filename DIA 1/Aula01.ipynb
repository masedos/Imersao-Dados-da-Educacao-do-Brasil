Desafio01: Proporção dos inscritos por idade.

Desafio03: Adicionar título no gráfico.

```
dados['NU_IDADE'].hist(bins=20);
plt.title('Distribuição das idade')
plt.legend('Idades')
plt.show()
```

```
dados['NU_NOTA_REDACAO'].hist(bins=20).set_title('Notas da Redação');
```

Desafio04: Plotar os Histogramas das idades dos do treineiro e não treineiros.

```
dados.query('IN_TREINEIRO == 1')['NU_IDADE'].hist(bins=20, figsize=(7, 5)) # Treineiro
dados.query('IN_TREINEIRO == 0')['NU_IDADE'].hist(bins=20, figsize=(7, 5)) # Não treineiro


plt.title('Treineiros X não treineiros no ENEM 2019');
plt.xlabel('Idades candidatos');
plt.ylabel('Quantidade de candidatos');
plt.legend(['Não treineiro', 'Treineiro']);
plt.show();
```

Desafio05: Comparar as distribuições das provas em inglês espanhol

```
dados.query('TP_LINGUA == 0')['NU_IDADE'].hist(bins=20, figsize=(7, 5))
dados.query('TP_LINGUA == 1')['NU_IDADE'].hist(bins=20, figsize=(7, 5), alpha=.6)

plt.title('Distribuições das provas em inglês espanhol no ENEM 2019');
plt.xlabel('Idades candidatos');
plt.ylabel('Quantidade de candidatos');
plt.legend(['Inglês', 'Espanhol']);
plt.show();
```
